```python
##Labelling

import os
import pandas as pd
df = pd.read_excel('C:/Users/Asus/OneDrive/Desktop/OASIS/OASIS-2/oasis_longitudinal_demographics-8d83e569fa2e2d30.xlsx')
data = []
root_dir = 'C:/Users/Asus/OneDrive/Desktop/OASIS/OASIS-2'


for mri_dir in os.listdir(root_dir):
    mri_path = os.path.join(root_dir, mri_dir, 'RAW')
    mri_id = mri_dir.split('_')[0] + '_' + mri_dir.split('_')[1]
    matched_rows = df[df['Subject ID'] == mri_id]
    if not matched_rows.empty:

        group = matched_rows['Subject ID'].values[0]
        if group == 'Converted':
            group = 'Demented'
        for file in os.listdir(mri_path):
            if file.endswith('.hdr'):
                file_path = os.path.join(mri_path, file)
                data.append({'File Path': file_path, 'Group': group})
    else:
        print(f"Warning: No matching MRI ID found for {mri_id}")
hdr_df = pd.DataFrame(data)
print(hdr_df.head())
```

```python
hdr_df.to_csv('Finally labelled.csv')
```
